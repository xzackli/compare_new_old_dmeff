```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

import numpy as np
import matplotlib.pyplot as plt
```

```python
ttcmb = np.load('output/dmeffcmb.npy')
ttkboddy = np.load('output/kboddy.npy')
fidkboddy = np.load('output/fidkboddy.npy')
fidcmb = np.load('output/fidcmb.npy')

plt.plot(ttkboddy / fidkboddy, "-", label="kboddy/class_public dmeff")
plt.plot(ttcmb / fidcmb, "-", label="DMeff-CMB")
plt.legend()
plt.title("1 GeV, baryons")
plt.ylabel(r"$(C_{\ell}^{\mathrm{dmeff}} / C_{\ell}^{\mathrm{fid}}) |_{\sigma_p = 2 \times 10^{-26} \, \mathrm{cm}^2}$")
```
